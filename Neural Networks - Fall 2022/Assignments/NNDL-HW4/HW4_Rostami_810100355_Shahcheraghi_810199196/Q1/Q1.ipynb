```python
#Turn data to array and then normalize it 
#Normalizing data with minmax method in dataframe to showcase

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))

values = df.values

normalized_data = scaler.fit_transform(values)

normalized_data
```

```python
#Calculating Correlation Table 

df1 = df[['PM2.5', 'PM2.5_Changping', 'PM2.5_Dingling', 'PM2.5_Dongsi', 'PM2.5_Guanyuan', 'PM2.5_Gucheng', 'PM2.5_Huairou', 'PM2.5_Nongzhanguan', 'PM2.5_Shunyi', 'PM2.5_Tiantan', 'PM2.5_Wanliu', 'PM2.5_Wanshouxigong']]

df1
corr_mat = df1.corr()

import seaborn as sns
plt.figure(figsize=(18,12))
heatmap = sns.heatmap(corr_mat, vmin=-1, vmax=1, annot=True,cmap=sns.cubehelix_palette(as_cmap=True))
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
```

```python
#Function for Turning data into supervised format

def supervise_data(data):

  lag_hour = 168

  X = []
  Y = []

  for i in range(lag_hour, data.shape[0] ):

    X.append(data[i-lag_hour:i,:])

    Y.append(data[i,0:1])
    
  return X,Y
```

```python

X, Y = supervise_data(normalized_data)

X = np.array(X)
Y = np.array(Y)


print('Y' ,Y.shape)
print('X' ,X.shape)

Y
```

```python
#Dividing data to test and train

X_train, X_test = X[:28052,] , X[28052:,]
print('X_train' ,X_train.shape)
print('X_test' ,X_test.shape)

Y_train, Y_test = Y[:28052,] , Y[28052:,]
print('Y_train' ,Y_train.shape)
print('Y_test' ,Y_test.shape)

Y_test

```

```python
#Building cnn-lstm model
import keras
from keras.preprocessing import sequence
from keras.models import Sequential
from keras.layers import Input,Dense, Dropout, Activation,Flatten, TimeDistributed, BatchNormalization
from keras.layers import LSTM
from keras.layers import Conv1D, MaxPooling1D
from __future__ import print_function


model = Sequential()            

model.add(Conv1D(filters=64, kernel_size=3, padding="causal", activation='relu',input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(BatchNormalization())

model.add(Conv1D(filters=64, kernel_size=3, padding="causal", activation='relu'))
model.add(BatchNormalization())

model.add(Conv1D(filters=32, kernel_size=3, padding="causal", activation='relu'))

model.add(MaxPooling1D(pool_size = 3))

model.add(LSTM(100, return_sequences=True))
model.add(Dropout(0.2))

model.add(LSTM(50))
model.add(Dropout(0.3))

model.add(Dense(1, activation="relu")) 

adam = keras.optimizers.Adam(learning_rate=0.0001, decay=0.00001)

model.compile(loss='mean_squared_error',optimizer= adam)

model.summary()

```

```python
#Training model

history =model.fit(X_train, Y_train,batch_size=32,epochs=200)
```

```python
plt.plot(history.history['loss'], label='Train loss')

plt.legend();
```

```python
#Predicting test data pm2.5 with trained model

Y_pred = model.predict(X_test)
```

```python


print('y_predicted',Y_pred.shape)

print('X_test' ,X_test.shape)
print('scaled Values shape', normalized_data.shape)
```

```python
#Reversing normalzed values to original format
d = normalized_data[:6844,:]
print('Y_pred',Y_pred.shape)
Y_predicted = np.concatenate((Y_pred,d[:6844,1:]), axis =1)
print('Concat Y pred',Y_pred.shape)
Y_tested = np.concatenate((Y_test, d[:6844,1:]), axis = 1)
print('Concat Y test', Y_test.shape)
```

```python
Y_predicted = scaler.inverse_transform(Y_predicted)
Y_tested = scaler.inverse_transform(Y_tested)
Y_predicted = Y_predicted[:,0:1]
Y_tested = Y_tested[:,0:1]
print('Y_tested', Y_tested.shape)
print('Y_predicted', Y_predicted.shape)
```

```python
#Calculating mae, rmse, r2

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

r2 = r2_score(Y_tested , Y_predicted)

mae = np.mean(np.abs(Y_tested - Y_predicted))

rmse = np.sqrt(mean_squared_error(Y_tested, Y_predicted))

print('mae:',mae)
print('RMSE:',rmse)
print('R2:',r2)
```
