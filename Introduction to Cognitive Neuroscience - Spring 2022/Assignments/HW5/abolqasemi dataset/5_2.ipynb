```python
fig, ax = plt.subplots()
spike_value = 1


# Loop to plot raster for each trial
for trial in range(len(n1['su'][0][0])):
    spike_times = [i for i, x in enumerate(n1['su'][0][0][trial]) if x == spike_value]
    ax.vlines(spike_times, trial - 0.5, trial + 0.5)

ax.set_xlim([0, len(n1['su'][0][0][0])])
ax.set_xlabel('Time (ms)')

fig.set_size_inches(10.5, 8.5)

ax.set_ylabel('Trial Number')
ax.set_title('Neuron 1 Raster plot') 


plt.show()
```

**PSTH**

```python
fig, axs = plt.subplots(2, 1, figsize=[10,10])
e = n1['su'][0][0]
e = np.delete(e,np.s_[:2500],axis=1)
e = np.delete(e,np.s_[5500-2500:],axis=1)


for trial in range(len(n1['su'][0][0])):
    spike_times = [i for i, x in enumerate(n1['su'][0][0][trial][2500:5500]) if x == spike_value]
    axs[0].vlines(spike_times, trial - 0.5, trial + 0.5)

    
axs[0].set_xlim([-1, len(e[0])])
axs[0].set_ylabel('Trial Number')
axs[0].set_title('Neuronal 1 Rater plot') 


# Draw the PSTH
axs[1].bar(range(e.shape[1]), np.mean(e, 0))

# Use same x axis limits as for raster plot
axs[1].set_xlim([-1, len(e[0])])
axs[1].set_ylim([0, 0.04])

axs[1].set_title('Peri-Stimulus Time Histogram (PSTH) of Neuron 1')
axs[1].set_xlabel('Time (ms)')
axs[1].set_ylabel('Probability of spike')
fig.tight_layout()
plt.show()
```
