**PSTH**

```python
cond = cond['Cond'].ravel()
for i in range(8):
    cond[np.where((cond==i+1) | (cond==i+1+8))[0]] = i
unique_labels = np.unique(cond)
print("labels:", unique_labels)
```

```python
n = [n1,n2,n3,n4,n5,n6,n7,n8,n9,n10,n11,n12,n13,n14,n15,n16]
su_dic = {}
list_of_su = [1, 2, 3, 5, 8, 9, 11, 13]
for i in list_of_su:
    mat = n[i-1]
    su_dic[str(i)] = mat['su'][0][0][:, 2500:5500].astype(float)
```

```python
idx_str_neuron_list = ['1', '2', '3', '5', '8', '9', '11', '13']
su_cond_dic = {}
for idx in idx_str_neuron_list:
    su_cond_dic[idx] = {}
    for label in unique_labels:
        su_cond_dic[idx][label] = su_dic[idx][np.where(cond==label)[0], :]
```

```python
fig, axs = plt.subplots(8, 8, figsize=(20, 15))
for i, idx in enumerate(idx_str_neuron_list):
    for j, label in enumerate(unique_labels):
        axs[i, j].hist(np.where(su_cond_dic[idx][label]==1)[1] + 2500, bins=100, color='black')

cols = ['Cond {} - {}'.format(col+1, col+1+8) for col in range(8)]
rows = ['Neuron {}'.format(row) for row in idx_str_neuron_list]

pad = 5 # in points

for ax, col in zip(axs[0], cols):
    ax.annotate(col, xy=(0.5, 1), xytext=(0, pad),
                xycoords='axes fraction', textcoords='offset points',
                size='large', ha='center', va='baseline')

for ax, row in zip(axs[:,0], rows):
    ax.annotate(row, xy=(0, 0.5), xytext=(-ax.yaxis.labelpad - pad, 0),
                xycoords=ax.yaxis.label, textcoords='offset points',
                size='large', ha='right', va='center')

fig.tight_layout()
fig.subplots_adjust(wspace=0.3)
plt.show()
```

## firing rate

```python
su_cond_moving_dic = {}
stride = 20
win_size = 50
for idx in idx_str_neuron_list:
    su_cond_moving_dic[idx] = {}
    for label in unique_labels:
        su_cond_arr = su_dic[idx][np.where(cond==label)[0]]
        i = 0
        mean_kernel_moving_list = []
        while True:
            if (i*stride)+win_size > 3000:
                break
            mean_kernel_moving_list.append(np.mean(su_cond_arr[:, i*stride:(i*stride)+win_size], axis=1))
            # print(i*stride, (i*stride)+50)
            i += 1
        su_cond_moving_dic[idx][label] = np.array(mean_kernel_moving_list).T
```

### firing rates over time for different stimulus angles

```python
for idx in idx_str_neuron_list:
    plt.figure(figsize=(10, 6))
    for label in unique_labels:
        mean = np.mean(su_cond_moving_dic[idx][label], axis=0)
        plt.title(f"Neuron {idx}")
        plt.ylabel("1 / Firing rate", size="large")
        plt.xlabel("Time sample (ms)", size="large")
        plt.plot(np.arange(mean.shape[0]), mean, label = "Condition " + str(label+1) + "-" + str(label+1+8))
        plt.legend()
    plt.show()
```

### firing rate over time for the stimuli with most and least responsiveness

```python
for idx in idx_str_neuron_list:
    plt.figure(figsize=(10, 6))
    for label in unique_labels:
        mean = np.mean(su_cond_moving_dic[idx][label], axis=0)
        ci = 1.96 * np.std(su_cond_moving_dic[idx][label], axis=0) / np.sqrt(len(su_cond_moving_dic[idx][label]))
        plt.title(f"Neuron {idx}")
        plt.ylabel("1 / Firing rate", size="large")
        plt.xlabel("Time sample (ms)", size="large")
        plt.plot(np.arange(mean.shape[0]), mean, label = "Condition " + str(label+1) + "-" + str(label+1+8))
        plt.fill_between(np.arange(mean.shape[0]), mean-ci, mean+ci, alpha=0.2)
        plt.legend()
    plt.show()
```
